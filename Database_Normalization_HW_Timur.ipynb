## Create 
The CREATE TABLE statement is used to create a new table in a database.

```.py
CREATE TABLE table_name (
    column1 datatype,
    column2 datatype,
    column3 datatype,
   ....
);
```
The column parameters specify the names of the columns of the table.

The datatype parameter specifies the type of data the column can hold (e.g. varchar, integer, date, etc.).
## Select 
The SELECT statement is used to select data from a database.

The data returned is stored in a result table, called the result-set.

```.py
SELECT column1, column2, ...
FROM table_name;
```
Here, column1, column2, ... are the field names of the table I want to select data from. If I want to select all the fields available in the table, I can use the following syntax:

```.py
SELECT * FROM table_name;
```
##Insert
It is possible to write the INSERT INTO statement in two ways.

###First way(Insert data into the table)
The first way specifies both the column names and the values to be inserted:
```.py
INSERT INTO table_name (column1, column2, column3, ...)
VALUES (value1, value2, value3, ...);
```
If I am adding values for all the columns of the table, I do not need to specify the column names in the SQL query. However, I need to make sure the order of the values is in the same order as the columns in the table. The INSERT INTO syntax would be as follows:
```.py
INSERT INTO table_name
VALUES (value1, value2, value3, ...);
```
###Second way(Insert Data Only in Specified Columns)
It is also possible to only insert data in specific columns.

The following SQL statement will insert a new record, but only insert data in the "CustomerName", "City", and "Country" columns (CustomerID will be updated automatically):
```.py
INSERT INTO Customers (CustomerName, City, Country)
VALUES ('Cardinal', 'Stavanger', 'Norway');
```
