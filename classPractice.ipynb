```
class Person():
  def __init__(self,name,age):
    self.name :str = name
    self.age :int = age

  def display():
    return f"{self.name} is {self.age} years old"

class Student(Person):
  def __init__(self,name,age,grade):
    Person.__init__(self,name,age)
    self.grade :int = grade
  
  def displayStudent(self):
    return f"Student name - {self.name}, age -  {self.age}, grade - {self.grade}"
  
```

```
student1 = Student(name = "Juan",age = "17", grade = "12")
student1.displayStudent()
```

First roman converter (1-999)

```
# Old roman converter(1-999)
class Conventer:
    def IntToRoman(number):       # 1 - 999
        number = str(number)
        digit = 0
        romanN = ""
        step = 0
        for i in number:  # 5 6 3
            digit += 1
            if digit == 1 and len(number) >= 3:
                if int(i) <= 3:
                    romanN += "C" * int(i)
                if int(i) == 4:
                    romanN += "CD"
                if int(i) >= 5 and int(i) != 9:
                    romanN += "D{}".format("C" * (int(i) - 5))
                if int(i) == 9:
                    romanN += "CM"

            # tens
            if digit == 2:
                if int(i) <= 3:
                    romanN += "X" * int(i)
                if int(i) == 4:
                    romanN += "XL"
                if int(i) >= 5 and int(i) != 9:
                    romanN += "L{}".format("X" * (int(i) - 5))
                if int(i) == 9:
                    romanN += "XC"
                # if int(i) == 0:
                #     romanN += "X"
            # ones
            if digit == 3 or len(number) == 1:
                if int(i) <= 3:
                    romanN += "I" * int(i)
                if int(i) == 4:
                    romanN += "IV"
                if int(i) >= 5 and int(i) != 9:
                    romanN += "V{}".format("I" * (int(i) - 5))
                if int(i) == 9:
                    romanN += "IX"
        print(romanN)

Conventer.IntToRoman(8)
Conventer.IntToRoman(654)

```

```
# New roman converter(1-9999)
class New_Conventer:
    def IntToRoman(self, number):
        val = [1000, 900, 500, 400,100, 90, 50, 40,10, 9, 5, 4,1]
        syb = ["M", "CM", "D", "CD","C", "XC", "L", "XL","X", "IX", "V", "IV","I"]
        romanN = ''
        i = 0
        while  number > 0:
            for _ in range(number // val[i]):
                romanN += syb[i]
                number -= val[i]
            i += 1
        return romanN

print(New_Conventer().IntToRoman(654))
print(New_Conventer().IntToRoman(4783))
```
